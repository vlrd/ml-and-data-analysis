```python
value = 1000
def sample_exp(n):
    list_x = []
    for i in range(value):
        x=exp_rv.rvs(size=n)
        list_x.append(x.mean())
    return list_x


def graph_sample(list_x,n):
    plt.hist(list_x, density=True)
    plt.plot(sorted(list_x), sts.norm(loc=np.mean(list_x),scale = np.std(list_x)).pdf(sorted(list_x)))
    plt.show()
```

```python
for n in [5,10,50]:
    list_x = sample_exp(n)
    graph_sample(list_x,n)
```
