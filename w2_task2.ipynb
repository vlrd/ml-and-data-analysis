```python
import numpy as np
from scipy import linalg
%matplotlib inline
from matplotlib import pylab as plt


def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)

def create_array(param):
    result = []
    if not isinstance(param, list):
        return result

    count = len(param)
    for i in range(count):
        result.append([param[i]**n for n in range(count)])
    return np.array(result)

def create_vector(param):
    result=[]
    if not isinstance(param, list):
        return result
    
    count = len(param)
    for i in range(count):
        result.append(f(param[i]))
    return result
```

```python
f_x = [f(x) for x in range(16)]
plt.plot(range(16), f_x)
plt.show()
```

```python
param = [1,15]
arr_1=create_array(param)
vec_1=create_vector(param)
answer = linalg.solve(arr_1, vec_1)
print(answer)
f_x=answer

plt.plot(param, f_x)
plt.show()
```

```python
param = [1,8, 15]
arr_2=create_array(param)
vec_2=create_vector(param)
answer = linalg.solve(arr_2, vec_2)
print(answer)
f_x=answer

plt.plot(param, f_x)
plt.show()
```
